# Load and Clean the Dataset

```python
import warnings
warnings.filterwarnings("ignore")
```

## Load the Dataset

```python
import pandas as pd

df = pd.read_csv('source/API_ITA_DS2_en_csv_v2_3472313.csv')
df.head()
```

## Drop Unnecessary Columns

```python
df.drop(['Country Name', 'Country Code', 'Indicator Code','Unnamed: 65'], axis=1, inplace = True)
df.head()
```

## Filter GDP-based Indicators

```python
df = df[df['Indicator Name'].str.contains('(% of GDP)')]
```

```python
df.to_csv('source/API_IT_GDP.csv')
```

## Transpose the dataset to obtain years as a single column

```python
df = df.transpose()
df.columns = df.iloc[0] 
df = df[1:]
df
```

## Drop Missing Values

```python
df.shape
```

```python
df.dropna(thresh=30, axis=1, inplace = True)
df = df.iloc[10:]
```

```python
df.shape
```

## Save the dataset

```python
df.to_csv('source/API_IT_GDP.csv')
```

## Plot Time Series

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_indicator(ts, indicator):
    fig_name = 'images/' + indicator.replace('/', "") + '.png'
    xmin = np.min(ts.index)
    xmax = np.max(ts.index)
    plt.figure(figsize=(15,6))
    plt.plot(ts)
    plt.title(indicator)
    plt.xticks(np.arange(xmin, xmax+1, 1),rotation=45)
    plt.xlim(xmin,xmax)
    plt.grid()
    plt.savefig(fig_name)
    return fig_name
```

# Run the Experiment

```python
# import comet_ml at the top of your file
from comet_ml import Experiment

# Create an experiment with your api key
experiment = Experiment()
```

```python
for indicator in df.columns:
    ts = df[indicator]
    ts.dropna(inplace=True)
    ts.index = ts.index.astype(int)
    fig = plot_indicator(ts,indicator)
    experiment.log_image(fig,name=indicator, image_format='png')

experiment.end()
```
